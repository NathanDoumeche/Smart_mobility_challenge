```python
import seaborn as sns
import pandas as pd
from pandas_profiling import ProfileReport

import warnings
warnings.filterwarnings("ignore", message="RuntimeWarning: overflow encountered in longlong_scalars")
```

```python
df_station = pd.read_csv("../0. Input Data/train.csv",sep=",") #v2
df_station['date']=pd.to_datetime(df_station['date'])
df_station.head()
```

# Use pandas-profiling
To get a first understanding of what is in the dataframe

```python
profile = ProfileReport(df_station, title="Pandas Profiling Report")
profile.to_widgets()
```

# Investigation of observations made within pandas-profiling : 

We observe that: 
- Dates do not all have the same number of entries
- Stations do not all have the same number of entries

What are the stations for which we have more/less data ?

```python
number_of_lines_per_station = df_station.groupby('Station').size().sort_values()

fig, ax = sns.utils.plt.subplots(figsize=(8, 14))
sns.barplot(x=number_of_lines_per_station, y=number_of_lines_per_station.index, color='grey', orient='h', ax=ax)
ax.set_xlabel('Number of lines')
ax.set_ylabel('Station')
fig.suptitle('Number of lines in the database per station')
fig.tight_layout()
sns.utils.plt.show()
```

Are there stations and/or periods where we have more data ?

```python
n_observations_per_week_per_station = df_station.set_index('date').groupby(['Station', pd.Grouper(freq='W')]).size().rename('Number of datapoints').reset_index()
n_observations_per_week_per_station['date'] = n_observations_per_week_per_station['date'].dt.strftime("%Y-%m-%d")
n_observations_per_week_per_station = n_observations_per_week_per_station.pivot(index='Station', columns='date', values='Number of datapoints').fillna(0)

fig, ax = sns.utils.plt.subplots(figsize=(12, 10))
sns.heatmap(n_observations_per_week_per_station, ax=ax)
fig.suptitle('Number of datapoints per station and per week')
fig.tight_layout()
sns.utils.plt.show()

# We create this arbitrary filter that will be used in other plots to hide irrelevant statistics when there is insufficient data for this station/week
mask_sufficient_data_per_station_per_week = n_observations_per_week_per_station < 400
```

# Evolution over time at station level

```python
for val in ['Available', 'Charging', 'Passive', 'Other']:
    average_per_week_per_station = df_station.set_index('date').groupby(['Station', pd.Grouper(freq='W')])[val].mean().rename(f"Average weekly {val}").reset_index()
    average_per_week_per_station['date'] = average_per_week_per_station['date'].dt.strftime("%Y-%m-%d")
    average_per_week_per_station = average_per_week_per_station.pivot(index='Station', columns='date', values=f"Average weekly {val}").fillna(0)

    cmap = sns.color_palette("rocket", as_cmap=True).copy()
    cmap.set_bad('#ccffcc')

    fig, ax = sns.utils.plt.subplots(figsize=(12, 10))
    sns.heatmap(
        data=average_per_week_per_station,
        mask=mask_sufficient_data_per_station_per_week,
        ax=ax,
        cmap=cmap,
        vmin=0,
        vmax=3,
    )
    fig.suptitle(f"Average weekly {val}")
    fig.tight_layout()
    sns.utils.plt.show()

```

```python
for val in ['Available', 'Charging', 'Passive', 'Other']:
    variability_per_week_per_station = df_station.set_index('date').groupby(['Station', pd.Grouper(freq='W')])[val].std().rename(f"Weekly {val} variability").reset_index()
    variability_per_week_per_station['date'] = variability_per_week_per_station['date'].dt.strftime("%Y-%m-%d")
    variability_per_week_per_station = variability_per_week_per_station.pivot(index='Station', columns='date', values=f"Weekly {val} variability").fillna(0)

    cmap = sns.color_palette("rocket", as_cmap=True).copy()
    cmap.set_bad('#ccffcc')

    fig, ax = sns.utils.plt.subplots(figsize=(12, 10))
    sns.heatmap(
        data=variability_per_week_per_station,
        mask=mask_sufficient_data_per_station_per_week,
        ax=ax,
        cmap=cmap,
        vmin=0,
        vmax=1.6,
    )
    fig.suptitle(f"Weekly {val} variability")
    fig.tight_layout()
    sns.utils.plt.show()
```

### Instead of looking at the full timelime, can we look at "time of day" and "day of week" ?

```python
sns.heatmap(df_station.groupby(['Station', 'tod'])['Available'].mean().unstack(), cmap=cmap)
sns.utils.plt.show()

sns.heatmap(df_station.groupby(['Station', 'dow'])['Available'].mean().unstack(), cmap=cmap)
sns.utils.plt.show()
```

### Figures above could not be interpreted : variability inter-station is too high compared to variability intra-station

### Remake the same figures, but normalize according to that station's typical availability

```python
toto = df_station.groupby(['Station', 'tod'])['Available'].mean().reset_index()
toto['Available'] = toto['Available'] / toto['Station'].map(df_station.groupby(['Station'])['Available'].mean())
sns.heatmap(toto.pivot(index='Station', columns='tod', values='Available'), vmin=0.7, vmax=1.3)
sns.utils.plt.show()


toto = df_station.groupby(['Station', 'dow'])['Available'].mean().reset_index()
toto['Available'] = toto['Available'] / toto['Station'].map(df_station.groupby(['Station'])['Available'].mean())
sns.heatmap(toto.pivot(index='Station', columns='dow', values='Available'), vmin=0.7, vmax=1.3)
sns.utils.plt.show()
```

### This previous analysis was conducted at station level. Are there interesting patterns at area level ?

# Evolution over time at area level

```python
for val in ['Available', 'Charging', 'Passive', 'Other']:
    average_per_week_per_station = df_station.set_index('date').groupby(['area', pd.Grouper(freq='W')])[val].mean().rename(f"Average weekly {val}").reset_index()
    average_per_week_per_station['date'] = average_per_week_per_station['date'].dt.strftime("%Y-%m-%d")
    average_per_week_per_station = average_per_week_per_station.pivot(index='area', columns='date', values=f"Average weekly {val}").fillna(0)

    cmap = sns.color_palette("rocket", as_cmap=True).copy()
    cmap.set_bad('#ccffcc')

    fig, ax = sns.utils.plt.subplots(figsize=(12, 4))
    sns.heatmap(
        data=average_per_week_per_station,
        ax=ax,
        cmap=cmap,
        vmin=0,
        vmax=3,
    )
    fig.suptitle(f"Average weekly {val}")
    fig.tight_layout()
    sns.utils.plt.show()
```

```python
for val in ['Available', 'Charging', 'Passive', 'Other']:
    variability_per_week_per_station = df_station.set_index('date').groupby(['area', pd.Grouper(freq='W')])[val].std().rename(f"Weekly {val} variability").reset_index()
    variability_per_week_per_station['date'] = variability_per_week_per_station['date'].dt.strftime("%Y-%m-%d")
    variability_per_week_per_station = variability_per_week_per_station.pivot(index='area', columns='date', values=f"Weekly {val} variability").fillna(0)

    cmap = sns.color_palette("rocket", as_cmap=True).copy()
    cmap.set_bad('#ccffcc')

    fig, ax = sns.utils.plt.subplots(figsize=(12, 4))
    sns.heatmap(
        data=variability_per_week_per_station,
        ax=ax,
        cmap=cmap,
        vmin=0,
        vmax=1.6,
    )
    fig.suptitle(f"Weekly {val} variability")
    fig.tight_layout()
    sns.utils.plt.show()
```

```python
sns.heatmap(df_station.groupby(['area', 'tod'])['Available'].mean().unstack(), cmap=cmap)
sns.utils.plt.show()

sns.heatmap(df_station.groupby(['area', 'dow'])['Available'].mean().unstack(), cmap=cmap)
sns.utils.plt.show()
```

```python
sns.heatmap(df_station.groupby(['area', 'tod'])['Available'].std().unstack(), cmap=cmap)
sns.utils.plt.show()

sns.heatmap(df_station.groupby(['area', 'dow'])['Available'].std().unstack(), cmap=cmap)
sns.utils.plt.show()
```
